## Treanding items analysis

```python
data_merged.order_date
```

```python
final_df=data_merged.dropna(subset=['order_date'])
```

```python
import pickle
pickle.dump(final_df,open('model/merge_item_orderitem_address.pkl','wb'))
```

```python
final_df.info()
```

```python
def items_data():
  today = date.today()
  today=pd.to_datetime(today)
  final_df['latest']=today-final_df.order_date
  new_items= final_df[["id","item_name","market","price","avgrating",'id','images','dealprice','type','discount','discountpercent','latest','state','qty']].sort_values(by=['latest'],ascending=True)
  return new_items
```

```python
items_df=items_data()
```

```python
items_df.latest
```

```python
df_final=items_df.loc[(items_df.latest<='100 days')]
```

```python
df_final.shape
```

```python
num_qty_df = df_final.groupby('item_name').count()['qty'].reset_index()
num_qty_df.rename(columns={'qty':'num_qty'},inplace=True)
num_qty_df
```

```python
data_final=df_final.merge(num_qty_df,on='item_name')
data_final.head()
```

```python
def trending_items(location):
  g=data_final.groupby(['state'])
  data_state=g.get_group(location)
  final_item=data_state[['avgrating','item_name','price','num_qty']]
  trending_item=final_item.sort_values(by='num_qty',ascending=False)
  trending_item.drop_duplicates('item_name',inplace=True)
  return trending_item.head()
```

```python
trending_items('Telangana')
```

```python
trending_items('Andhra Pradesh')
```

```python
trending_items('West Bengal')
```

new_arrival

```python
def new_arrival():
  today = date.today()
  today=pd.to_datetime(today)
  final_df['latest']=today-final_df.createdat
  new_items= final_df[["id","item_name","market","price","avgrating",'id','images','dealprice','type','discount','discountpercent','latest','state','qty']].sort_values(by=['latest'],ascending=True)
  return new_items
```

```python
new_arrival()
```

```python
def best_pick(state,item):
  data=new_arrival()
  new_product_state=data.groupby(['state'])
  new_prd=new_product_state.get_group(state)
  g1=new_prd.groupby(['market'])
  df_item=g1.get_group(item)
  final_item=df_item[["item_name","market","price","avgrating",'id','images','dealprice','type','discount','discountpercent','latest','state']]
  best_pick_item=final_item.sort_values(by='latest',ascending=True)
  best_pick_item.drop_duplicates('item_name',inplace=True)
  return best_pick_item.head()
```

```python
best_pick('Telangana','Clothes & Fashion')
```
